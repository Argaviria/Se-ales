```python
#Importamos las librerias 

import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#cargamos las imagenes
imagen1 = plt.imread("waste1.jpg")
imagen2 = plt.imread("waste2.jpg")
imagen3 = plt.imread("waste3.jpg")
imagen4 = plt.imread("waste4.jpg")

#despliega las imagenes para saber cuáles son las funciones con las que vamos a operar
#Despliega además el tamaño de las imágenes para saber si es posible operar entre ellas si hay que reescalarlar de modo que sean operables entre sí

#Se muestra el tamaño de la imagen "waste1.jpg"
print("waste1.jpg (Tamaño)" + str(imagen1.shape))
#despliega la imagen "waste1.jpg"
plt.imshow(imagen1)
plt.show()

#Se muestra el tamaño de la imagen "waste2.jpg"
print("waste2.jpg (Tamaño)" + str(imagen2.shape))
#despliega la imagen "waste1.jpg"
plt.imshow(imagen2)
plt.show()

#Se muestra el tamaño de la imagen "waste3.jpg"
print("waste3.jpg (Tamaño)" + str(imagen3.shape))
plt.imshow(imagen3)
plt.show()

#Se muestra el tamaño de la imagen "waste4.jpg"
print("waste4.jpg (Tamaño)" + str(imagen4.shape))
plt.imshow(imagen4)
plt.show()
```

```python
#Reescalamos las imágenes de modo que queden todas de la misma dimensión para que puedan sumarse entre ellas, 
#esto se hace con la función "cv2.MakeBorder" que crea un borde que delimita un nuevo tamaño para la imagen
#se hace para todas las imágenes y se despliegan las nuevas imágenes 

image1a = cv2.copyMakeBorder(imagen1, 43, 45, 84, 84, cv2.BORDER_CONSTANT)
print("waste1.jpg (Tamaño)" + str(image1a.shape))
plt.imshow(image1a)
plt.show()

image2a = cv2.copyMakeBorder(imagen2, 0, 0, 60, 60, cv2.BORDER_CONSTANT)
print("waste2.jpg (Tamaño)" + str(image2a.shape))
plt.imshow(image2a)
plt.show()

image3a = cv2.copyMakeBorder(imagen3, 0, 0, 60, 60, cv2.BORDER_CONSTANT)
print("waste3.jpg (Tamaño)" + str(image3a.shape))
plt.imshow(image3a)
plt.show()

image4a = cv2.copyMakeBorder(imagen4, 10, 9, 0, 1, cv2.BORDER_CONSTANT)
print("waste2.jpg (Tamaño)" + str(image2a.shape))
plt.imshow(image2a)
plt.show()
```

Suma

```python
# Aplicamos la operación suma sabiendo ahora que todas las imágenes son del mismo tamaño
# Primero se aplica la operación a todas las imagágenes pero no tiene mucho sentido, pues se superpondrían como ruido
fimg = (image1a + image2a + image3a + image4a)

#guardamos la imagen de resultado
cv2.imwrite('output.jpg', fimg)
plt.imshow(fimg)
```

```python
# Aplicamos ahora la superposición de las dos primeras imágenes 
fsimg = cv2.add(image1a,image2a)

#guardamos la imagen de resultado
cv2.imwrite('output.jpg', fsimg)
plt.imshow(fsimg)
```

```python
# Aplicamos ahora la superposición de las dos primeras imágenes cambiando el orden para probar si cambia la respuesta final 
fsiimg = cv2.add(image2a,image1a)

#guardamos la imagen de resultado
cv2.imwrite('output.jpg', fsiimg)
plt.imshow(fsiimg)
```

Resta

```python
# Aplicamos la operación resta sabiendo ahora que todas las imágenes son del mismo tamaño
fmimg = image1a - image2a - image3a - image4a

#guardamos la imagen de resultado
cv2.imwrite('output.jpg', fmimg)
plt.imshow(fmimg)
```

```python
# Aplicamos ahora la resta de las dos primeras imágenes 
frimg = cv2.subtract(image1a,image2a)

#guardamos la imagen de resultado
cv2.imwrite('output.jpg', frimg)
plt.imshow(frimg)
```

```python
# Aplicamos ahora la resta de las dos primeras imágenes cambiando el orden
friimg = cv2.subtract(image2a,image1a)

#guardamos la imagen de resultado
cv2.imwrite('output.jpg', friimg)
plt.imshow(friimg)
```

Multiplicación


```python
#hacemos la operación de multiplicación por un escalar para incrementar la iluminación
fmulimg = image3a * 0.02

#guardamos la imagen de resultado
cv2.imwrite('output.jpg', fmulimg)
plt.imshow(fmulimg)
```

```python
#usando la función de open CV para la multiplicación procedemos de la misma forma 
fmulimg = cv2.multiply(image3a, 0.02)

#guardamos la imagen de resultado
cv2.imwrite('output.jpg', fmulimg)
plt.imshow(fmulimg)
```

División


```python
#hacemos la operación de división por un escalar para incrementar la iluminación
fdivimg = image4a / 2

#guardamos la imagen de resultado
cv2.imwrite('output.jpg', fdivimg)
plt.imshow(fdivimg)
```

```python
#usando la función de open CV para la multiplicación procedemos de la misma forma 
fdivimg = cv2.divide(image4a, 2)

#guardamos la imagen de resultado
cv2.imwrite('output.jpg', fdivimg)
plt.imshow(fdivimg)
```
